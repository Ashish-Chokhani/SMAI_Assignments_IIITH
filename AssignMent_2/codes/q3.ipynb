```python
#---------------3.1----Hierarchial clustering and linkage matrix-----------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.preprocessing import StandardScaler

class HierarchicalClustering:
    def __init__(self):
        self.linkage_matrix = None

    def compute_linkage_matrix(self, data, linkage_method='ward'):
        self.linkage_matrix = linkage(data, method=linkage_method)

    def plot_dendrogram(self):
        if self.linkage_matrix is None:
            raise ValueError("Linkage matrix is not computed. Use compute_linkage_matrix method first.")
        
        plt.figure(figsize=(10, 5))
        dendrogram(self.linkage_matrix)
        plt.title('Hierarchical Clustering Dendrogram')
        plt.xlabel('Sample Index')
        plt.ylabel('Distance')
        plt.show()

    def perform_clustering(self, data, linkage_method='ward'):
        self.compute_linkage_matrix(data, linkage_method)
        self.plot_dendrogram()

if __name__ == "__main__":
    data = pd.read_csv('SMAI-Dataset-hc-dataset/new_customers.csv')
    data = data.drop(data.columns[0],axis=1)
    data.iloc[:, 0] = data.iloc[:, 0].map({'Male': 1, 'Female': 0})
    scaler = StandardScaler()
    data.iloc[:,1:] = scaler.fit_transform(data.iloc[:,1:])
    linkage_methods = ['ward', 'single', 'complete', 'average']
    feature_subsets = [[1, 2], [0, 2], [0, 1, 2], [0, 1, 2, 3]]  # Specify different feature subsets

    for linkage_method in linkage_methods:
        for feature_subset in feature_subsets:
            hc = HierarchicalClustering()
            selected_features_data = data.iloc[:, feature_subset]
            print(f"Linkage: {linkage_method}, Features Used: {feature_subset}")
            hc.perform_clustering(selected_features_data, linkage_method)
```

Varying the linkage matrix and features affects the shape and number of clusters discovered.Varying the features i.e. selecting a subset of the original features to use for analysis help to reduce noise and improve the interpretability of the results.
